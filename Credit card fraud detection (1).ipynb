```python
#Here we have very less data point which is fradulant, so we cannot use this to tain the model
#So we have to use the process of handling unbalanced data
#Here in Class label 0 indicate normal transaction and 1 indicate fraud transaction.

#Seperating data for analysis
```

```python
genuine = data[data.Class==0]
fraud = data[data.Class==1]
```

```python
# Now since we have unbalanced data so we will build a sample data from the original data such that it contains similar number
#genuine transaction and fraud transaction. We will takd all the fraud datapoint 
```

```python
genuine_sample = genuine.sample(n=492) #it will randomly sample the data
```

```python
new_data=pd.concat([genuine_sample, fraud], axis=0) #it will be added row wise 
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x=new_data.drop(columns='Class', axis=1)
y=new_data['Class']
```

```python
x.head()
```

```python
y.head()
```

```python
#Now we will split the data into training data and testing data
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
#stratify = y means that there will be evenly distributed data of 'class' in test and train
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
print(y_test)
```

Model Training

Linear Regression

```python
model=LogisticRegression()
```

```python
#Now we will train the model.
```

```python
model.fit(x_train, y_train)
```

```python
#Checking the model with accuracy score
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
```

```python
training_data_accuracy
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction, y_test)
```

```python
test_data_accuracy
```

```python
#if trainning accuracy is very high for example train_accuracy =94% and test_accuracy=50% then there is overfitting of
#training data and vice versa for underfitting.
```
